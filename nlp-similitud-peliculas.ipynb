```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

# 1. procesar el texto

```python
# Preprocesamiento de texto
stop_words = set(stopwords.words('english'))  # stopwords en inglés
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Eliminar caracteres especiales y números
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convertir a minúsculas
    text = text.lower()
    # Tokenización
    tokens = word_tokenize(text)
    # Eliminar stopwords 
    tokens = [word for word in tokens if word not in stop_words]
    # Unir tokens en una cadena de texto nuevamente
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text
```

```python
df['Processed_Plot'] = df['Plot'].apply(preprocess_text)
df.head()
```

## 2. Vectorización de palabras (word embeddings) 

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import pandas as pd


# Tokenización de las sinopsis
tokenized_plots = [word_tokenize(plot) for plot in df['Processed_Plot']]

# Entrenamiento del modelo Word2Vec
embedding_size = 100  # Tamaño del espacio de embeddings
window_size = 5  # Tamaño de la ventana de contexto
min_word_count = 1  # Frecuencia mínima de palabra para ser considerada
sample = 1e-3  # Umbral de submuestreo
sg = 1  # Usar skip-gram

model = Word2Vec(sentences=tokenized_plots,
                 vector_size=embedding_size,
                 window=window_size,
                 min_count=min_word_count,
                 sample=sample,
                 sg=sg)

# Entrenamiento completado, ahora puedes utilizar los embeddings

# Ejemplo: Obtener el embedding para una palabra específica
word_embedding = model.wv['movie']
print("Embedding para la palabra 'movie':", word_embedding)

# Guardar el modelo entrenado para su uso posterior
model.save("movie_embeddings.model")
```

## 3. Calcular la similitud de las sinopsis con similitud coseno

```python
from sklearn.metrics.pairwise import cosine_similarity

# Obtener los embeddings de las sinopsis de las películas
embeddings = []

for plot in df['Processed_Plot']:
    plot_embedding = []
    for word in plot.split():
        if word in model.wv.key_to_index:
            plot_embedding.append(model.wv[word])
    if plot_embedding:
        embeddings.append(np.mean(plot_embedding, axis=0))

# Calcular la similitud coseno entre todos los pares de embeddings
similarity_matrix = cosine_similarity(embeddings, embeddings)

# Imprimir la matriz de similitud
print("Matriz de Similitud:")
print(similarity_matrix)
```

```python
print("Tamaño de similarity_matrix:", similarity_matrix.shape)
```

## 4. Recomendar peliculas a partir de su nombre y de la similitud coseno entre los embedings de cada sinopsis.

```python
def recomendar_peliculas(idx_pelicula, similarity_matrix, df, top_n=20):
    # Obtener las similitudes de la película de entrada con todas las demás películas
    sim_scores = list(enumerate(similarity_matrix[idx_pelicula]))
    #print(sim_scores)

    # Ordenar las similitudes
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Obtener los índices de las películas más similares (excluyendo la película de entrada)
    sim_scores = sim_scores[1:top_n+1]  # Excluyendo la película de entrada

    # Obtener los índices y títulos de las películas recomendadas
    movie_indices = [i[0] for i in sim_scores]
    recommended_movies = df.iloc[movie_indices]['Title'].values

    return recommended_movies
```

```python
idx_pelicula_entrada = 60  
peliculas_recomendadas = recomendar_peliculas(idx_pelicula_entrada, similarity_matrix, df)

print(f"Películas recomendadas para la película de : {df.loc[idx_pelicula_entrada,'Title']}:")
for idx, movie in enumerate(peliculas_recomendadas, 1):
    print(f"{idx}. {movie}")
```
